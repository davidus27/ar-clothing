```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
import svgwrite
import os
import random
```

```python
# PATHS 
## Fonts
japanese_font_path = "fonts/SourceHanSerif-VF.ttf"
icon_font_path = "fonts/Symbola.ttf"

## Images output
mockup_path = "generated_mockups/"
```

```python
# Define the symbols
symbols = ["âšª", "âœ–", "â–²", "â– ", "â˜…", ">", "=", "â†’"]

# Japanese word (single character)
japanese_word = "å…‰"

# Create a new blank image with black background
img_width, img_height = 800, 400  # Adjusted width for the layout
background_color = (0, 0, 0)  # Black background
img = Image.new("RGB", (img_width, img_height), background_color)
draw = ImageDraw.Draw(img)

# Load a bold font for the Japanese character
try:
    font_size_japanese = 300  # Larger font size for the Japanese character
    font_japanese = ImageFont.truetype(japanese_font_path, font_size_japanese)
except IOError:
    print("Font file not found. Using default font.")
    font_japanese = ImageFont.load_default()  # Fallback to default font

# Draw the Japanese word on the left side
# Use textbbox instead of textsize
bbox = draw.textbbox((0, 0), japanese_word, font=font_japanese)
text_width = bbox[2] - bbox[0]
text_height = bbox[3] - bbox[1]

draw.text((50, 0), japanese_word, fill=(255, 255, 255), font=font_japanese)  # White text

# Load a smaller font for the icons
try:
    font_size_icon = 62  # Smaller font size for the icons
    font_icon = ImageFont.truetype(icon_font_path, font_size_icon)
except IOError:
    print("Font file not found. Using default font.")
    font_icon = ImageFont.load_default()  # Fallback to default font

# Define grid properties for icons
grid_size = (4, 2)  # 4 columns and 2 rows
icon_padding = 30  # Space between icons

# Calculate the grid position (to the right of the Japanese character)
grid_start_x = 400  # Start drawing the grid at this x-coordinate
# Center the grid vertically based on the height of the Japanese character
grid_start_y = (img_height) / 3  # Center the grid


# Draw the icons in a grid
for row in range(grid_size[1]):
    for col in range(grid_size[0]):
        index = row * grid_size[0] + col
        if index < len(symbols):
            icon = symbols[index]
            # Calculate position
            x = grid_start_x + col * (font_size_icon + icon_padding)
            y = grid_start_y + row * (font_size_icon + icon_padding)
            draw.text((x, y), icon, fill=(255, 255, 255), font=font_icon)  # Use icon font

# Save the image
output_path = mockup_path + "design_image.png"
img.save(output_path)

# Display the image using matplotlib
plt.imshow(img)
plt.axis('off')  # Hide axis
plt.show()

print(f"Image saved as {output_path}.")
```
